```python
import math

n = 10000

x=n
if x<6:
    x = 6

listofnum = [i for i in range(2,int(x*(math.log(x)+math.log(math.log(x)))))]

currentnumber = 0

for i in range(0,n):

    currentnumber = listofnum[0]
    for j in listofnum:
        if j % currentnumber == 0:
            listofnum.remove(j)

print(currentnumber)
```
